```python
import wntr

# 创建一个新的供水网络模型
net = wntr.network.WaterNetworkModel('net14test1.inp')

# 设置时间参数
time = wntr.sim.EpanetSimulator(net,reader="net14test1.inp") #.Timestep(start=0, duration=86400)

# 设置边界条件，例如水库的初始水位和流量
boundary = net..boundaries
boundary.head = 100.0  # 单位：米
boundary.demand_pattern = net.create_pattern('Constant', value=0.0)

# 运行模拟
sim = wntr.sim.WaterNetworkSimulator(net, timesteps=[time])
sim.run_model()

# 查看模拟结果
for t in sim.get_times():
    print(f"Time: {t}, Total Demand: {sim.get_total_demand(t)}, Total Supply: {sim.get_total_supply(t)}")

# 可视化模拟结果（需要额外的可视化工具，如 matplotlib）
```

```python
import epynet as en
import subprocess
import logging
import os

# 配置日志
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# 创建一个新的网络
network = en.Network()

# 添加节点（结点）
junction1 = network.add_junction('J1',x=0 , y=0,basedemand=0,elevation=1)
junction2 = network.add_junction('J2',x=10 , y=0,basedemand=0,elevation=1)
reservoir = network.add_reservoir('R1',x=0 , y=0,elevation=10)

# 设置结点参数
junction1.elevation = 100.0  # 结点1的高程
junction2.elevation = 90.0   # 结点2的高程
# reservoir.head = 150.0       # 水源的水头

# 添加管道
pipe1 = network.add_pipe('P1', 'R1', 'J1')
pipe2 = network.add_pipe('P2', 'J1', 'J2')

# 设置管道参数
pipe1.length = 1000.0  # 管道1的长度
pipe1.diameter = 0.3   # 管道1的直径
pipe1.roughness = 0.01 # 管道1的粗糙度

pipe2.length = 1000.0  # 管道2的长度
pipe2.diameter = 0.3   # 管道2的直径
pipe2.roughness = 0.01 # 管道2的粗糙度

# 设置结点需求
junction1.base_demand = 100.0  # 结点1的基本需水量
junction2.base_demand = 100.0  # 结点2的基本需水量

# 运行稳态水力分析
# results = network.run()

# 启用日志功能

# 运行稳态水力分析
results = network.run()
# 保存 EPANET 输入文件
input_file = 'net14test1.inp'
output_file = 'net14test1.out'
report_file = 'net14test1.rpt'

# network.write_inpfile(input_file)

# 使用 subprocess 调用 EPANET 并捕获输出 , output_file, report_file
try:
    result = subprocess.run(
        ['epanet', input_file],
        capture_output=True,
        text=True,
        check=True
    )
    logger.info(result.stdout)
    if result.stderr:
        logger.error(result.stderr)
except subprocess.CalledProcessError as e:
    logger.error(f"EPANET 运行失败: {e}")
    logger.error(e.stderr)
    exit(1)

# 读取 EPANET 输出文件
results = en.read(output_file)
# 输出结果
if results is None:
    print("运行稳态水力分析失败，请检查输入数据和配置。")
else:
    print(results)
    
    # 使用 en.getnodevalue 和 en.getlinkvalue 获取详细结果
    print(f"结点 J1 的压力: {en.getnodevalue(results, 'J1', en.PRESSURE)} 米")
    print(f"结点 J2 的压力: {en.getnodevalue(results, 'J2', en.PRESSURE)} 米")
    print(f"管道 P1 的流量: {en.getlinkvalue(results, 'P1', en.FLOW)} 立方米/秒")
    print(f"管道 P2 的流量: {en.getlinkvalue(results, 'P2', en.FLOW)} 立方米/秒")
```
