```python
import networkx as nx
import matplotlib.pyplot as plt

# 网络图的基本信息
G = nx.DiGraph()  # 创建一个有向图对象

# 添加节点
nodes = ['A', 'B', 'C', 'D']
G.add_nodes_from(nodes)

# 添加边，同时添加边的属性（如摩擦系数）
edges = [('A', 'B', {'friction': 0.1}),
         ('A', 'D', {'friction': 0.5}),
         ('B', 'C', {'friction': 0.2}),
         ('C', 'D', {'friction': 0.3}),
         ('B', 'D', {'friction': 0.4})]
G.add_edges_from(edges)

# 设置节点位置
pos = {'A': (0, 0), 'B': (1, 1), 'C': (2, 0), 'D': (3, 1)}

# 绘制节点和边
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500)

# 绘制边上的标签（摩擦系数）
edge_labels = nx.get_edge_attributes(G, 'friction')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# 显示图形
plt.title('Network Graph Visualization')
plt.show()
```
